```sql
create or replace table dfa_sales_transactions as
select *
from read_files('/Volumes/workspace/dataforge_work/dataforge/dfa_sales_transactions.csv', format => 'csv', header => true);

describe dfa_sales_transactions;
```

```sql
--fix null values in Discount column
update dfa_sales_transactions
set Discount = 0
where Discount is null;

```

```sql
--Display all records
select * 
from dfa_sales_transactions
```

```sql
--show Product_name, Region, Quantity, Unit_price

select Product_Name, Region, Quantity, Unit_Price
from dfa_sales_transactions;
```

```sql
--Show all transactions from Guateng
select Transaction_ID, Region
from dfa_sales_transactions
where Region = 'Gauteng';
```

```sql
--show all transactions where quantity is greater than 2
select *
from dfa_sales_transactions
where Quantity > 2;
```

```sql
--show all transactions where payment_method is 'card'
select *
from dfa_sales_transactions
where Payment_Method = 'Card';
```

```sql
--Calculate the total revenue (quantity*unit_price)
select sum(Quantity*Unit_Price) as total_revenue
from dfa_sales_transactions;
```

```sql
--calculate the total quantity sold
select sum(Quantity) as total_quantity_sold
from dfa_sales_transactions
```

```sql
--find the average unit price
select AVG(Unit_Price) as avg_unit_price
from dfa_sales_transactions;
```

```sql
--calculate total revenue per Product_name
select Product_Name, sum(Quantity*Unit_Price) as total_revenue
from dfa_sales_transactions
group by Product_Name;
```

```sql
--calculate total revenue per Region
select Region, sum(Quantity*Unit_Price) as total_revenue
from dfa_sales_transactions
group by Region;
```

```sql
--identify the top selling product(based on quantity)
select Product_Name, sum(Quantity) as top_selling_product
from dfa_sales_transactions
group by Product_Name
order by top_selling_product desc;
```

```sql
--identify the best performing region(based on revenue)
select Region, sum(Quantity*Unit_Price) as total_revenue
from dfa_sales_transactions
group by Region
order by total_revenue desc;
```

```sql
--calculate revenue after discount: Revenue = Quantity * Unit_Price * (1 - Discount)
select Region, sum(Quantity*Unit_Price*(1-Discount/100)) as total_revenue
from dfa_sales_transactions
group by Region;
```

```sql
--which category generates the highest revenue?
select Category, sum(Quantity*Unit_Price) as total_revenue
from dfa_sales_transactions
group by Category
order by total_revenue desc;
```

```sql
--which payment is used the most?
select Payment_Method, count(*) as total_transactions
from dfa_sales_transactions
group by Payment_Method
order by total_transactions desc;
```
